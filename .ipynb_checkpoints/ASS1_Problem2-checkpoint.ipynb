```python
import pandas as pd
import numpy as np
```

```python
Dataset = pd.read_excel('CoffeeShopTransactions.xlsx')
Dataset_copy = Dataset.copy()

#copy of Data_set
Dataset_copy
```

```python
Dataset_copy.info() 
```

# Data Cleaning

```python
Dataset_copy = Dataset_copy.replace('Caramel bites', 'CaramelBites') 

# We find Caramel bites and CaramelBites and they are the same item but with different writing
```

```python
Trans_ID= Dataset_copy["Transaction Number"].values
Trans_ID
```

```python
ITEMS= Dataset_copy[["Item 1","Item 2","Item 3"]].values
ITEMS
```

```python
MergedData=[]
for i in range(ITEMS.shape[0]):
    MergedData.append([Trans_ID[i],ITEMS[i]])
MergedData
```

```python
DistinctItems = []
for i in MergedData:
    for q in i[1]:
        if(q not in DistinctItems): #add distinct ITEMS
            DistinctItems.append(q)
DistinctItems = sorted(DistinctItems)
print(DistinctItems)
```

#  input ---> Support ---> Confidence --- 

```python
Min_Support = float(input("Enter Min Support: "))
Min_Support_Count = int(Min_Support*ITEMS.shape[0])

Min_Confidence = float(input("Enter Min Confidence: "))

Min_Support_Count
```

# Alpriori Algorithm

```python
#Map={}
```

```python
from collections import Counter
AllWordsCount = Counter()
for i in DistinctItems:
    for Line in MergedData:
        if(i in Line[1]):
            AllWordsCount[i]+=1
print("Words 1:")
for i in AllWordsCount:
    print(str([i])+": "+str(AllWordsCount[i]))
print()
WordsSatisfied = Counter()
for i in AllWordsCount:
    if(AllWordsCount[i] >= Min_Support_Count):
        WordsSatisfied[frozenset([i])]+=AllWordsCount[i]
print("Words Satisfied 1:")
for i in WordsSatisfied:
    #Map[str(list(i))]=str(WordsSatisfied[i])
    print(str(list(i))+": "+str(WordsSatisfied[i]))
print()
```

```python
#Map
```

```python
LastWordsSatisfied = WordsSatisfied # the LastWordsSatisfied we get association rules from it
pos = 1 # position of last words Satisfied 
for count in range (2,1000): # as we get the first words satisfied in the above cell  "count needed for num_Items"
    AllWordsSet = set()
    temp = list(WordsSatisfied)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            WordsCombination = temp[i].union(temp[j])
            #print(frozenset(WordsCombination))       # in : brownie , cake ---> {brownie , cake}
            if (len(WordsCombination)==count):
                AllWordsSet.add(WordsCombination)
    AllWordsSet = list(AllWordsSet) #has all combinations of the L
    AllWordsCount = Counter() #Get Words Set COunt
    for i in AllWordsSet:
        AllWordsCount[i] = 0
        for Line in MergedData:
            ItemsSet = set(Line[1])
            if(i.issubset(ItemsSet)):
                AllWordsCount[i]+=1
    print("Words "+str(count)+":")
    for i in AllWordsCount:
        print(str(list(i))+": "+str(AllWordsCount[i]))
    print()
    WordsSatisfied = Counter() #Get that Satisfy COndition >= Min Support Count Calculated
    for i in AllWordsCount:
        if(AllWordsCount[i] >= Min_Support_Count):
            WordsSatisfied[i]+=AllWordsCount[i]
    print("Words Satisfied "+str(count)+":")
    for i in WordsSatisfied:
        #Map[str(list(i))]=str(WordsSatisfied[i])
        print(str(list(i))+": "+str(WordsSatisfied[i]))
    if(len(WordsSatisfied)==0):
        print("....THERE IS NO WORDS SATISFIED....")
    print()
    if(len(WordsSatisfied) == 0):
        break
    LastWordsSatisfied = WordsSatisfied
    pos = count
print("Result: ")
print("Last Words Satisfied "+str(pos)+":")
for i in LastWordsSatisfied:
    print(str(list(i))+": "+str(LastWordsSatisfied[i]))
print()
```

# Association rules for the Subsets

```python
from itertools import combinations
StrongRules=set()
for l in LastWordsSatisfied:
    if(len(l)==1): #as there is no frequent terms together
            break
    WholeRule = [frozenset(q) for q in combinations(l,len(l)-1)]
    for PartRule in WholeRule:
        RemainingRule = l-PartRule
        Rule = l
        #SupportRule = int(Map[str(list(Rule))])
        #SupportPartRule = int(Map[str(list(PartRule))])
        #SupportRemainingRule =int(Map[str(list(RemainingRule))])
        SupportRule = 0
        SupportPartRule = 0
        SupportRemainingRule = 0
        for q in MergedData:
            ITEMSLINE = set(q[1])
            if(PartRule.issubset(ITEMSLINE)):
                SupportPartRule+=1
            if(RemainingRule.issubset(ITEMSLINE)):
                SupportRemainingRule+=1
            if(Rule.issubset(ITEMSLINE)):
                SupportRule+=1
        #print(Map[str(list(PartRule))])
        #print(Map[str(list(RemainingRule))])
        #print(Map[str(list(l))])
        
        ConfidenceCalculated = SupportRule/SupportPartRule*100
        if(ConfidenceCalculated > Min_Confidence*100):
            print("This Next Association Rule is Strong")
            StrongRules.add(str(list(PartRule))+" -> "+str(list(RemainingRule))+" = "+str(SupportRule/SupportPartRule*100)+"%")
        ConfidenceCalculated = SupportRule/SupportRemainingRule*100
        print(str(list(PartRule))+" -> "+str(list(RemainingRule))+" = "+str(SupportRule/SupportPartRule*100)+"%")
        print()
        if (len(WholeRule)>2): # to handle Duplicated Data Rules
            if(ConfidenceCalculated >= Min_Confidence*100):
                print("This Next Association Rule is Strong")
                StrongRules.add(str(list(RemainingRule))+" -> "+str(list(PartRule))+" = "+str(SupportRule/SupportRemainingRule*100)+"%")
            print(str(list(RemainingRule))+" -> "+str(list(PartRule))+" = "+str(SupportRule/SupportRemainingRule*100)+"%")
            print()
        
    print()
    print()
```
